```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 10, 5
#plt.style.use('fivethirtyeight')
plt.style.use('bmh')
```

```python
df = pd.read_excel("../Extraction/Yields_Interpolated.xlsx", index_col=0)
df1 = pd.read_excel("../Extraction/Yields_Interpolated.xlsx", index_col=0, sheetname="Sheet2")
```

```python
df.head()
```

```python
short = df.loc[:, :"3-Month WAV"].copy()
long = df.loc[:, :"10-Year WAV"].copy()
```

```python
short_descript = short.loc[:, :"3-Month WAV"].describe()
short_descript.to_excel("short_descript.xlsx")
```

```python
short[["3-Month DEU"]].plot()
df1["Recession"].plot()
plt.title("3-Month DEU")
plt.legend()
plt.savefig("../Graphs/3-Month DEU.png", dpi=100)
```

```python
short["3-Month WAV"].plot(color="g")
df1["Recession"].plot()
plt.title("3-Month WAV")
plt.legend()
plt.savefig("../Graphs/3-Month WAV.png", dpi=100)
```

```python
long_descript = long.loc[:, :"10-Year WAV"].describe()
long_descript.to_excel("long_descript.xlsx")
```

```python
long[["10-Year DEU"]].plot()
df1["Recession"].plot()
plt.title("10-Year DEU")
plt.legend()
plt.savefig("../Graphs/10-Year DEU.png", dpi=100)
```

```python
long["10-Year WAV"].plot(color="g")
df1["Recession"].plot()
plt.title("10-Year WAV")
plt.legend()
plt.savefig("../Graphs/10-Year WAV.png", dpi=100)
```
